```python
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
 matches["venue_code"] = matches["Venue"].astype("category").cat.codes 
```

```python
matches["opponent_code"] = matches["Opponent"].astype("category").cat.codes
```

```python
 matches["hour"] = matches["Time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["Date"].dt.dayofweek
```

```python
matches["target"] = (matches["Result"] == "W").astype("int")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["Date"] < '2022-01-01']
train
```

```python
test = matches[matches["Date"] > '2022-01-01']
test
```

```python
predictors = ["venue_code", "opponent_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test["target"], preds)
accuracy 
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
combined
```

```python
 grouped_matches = matches.groupby("Team")
```

```python
group = grouped_matches.get_group("Barcelona")
group
```

```python
def rolling_averages(group, columns, new_columns):
  group = group.sort_values("Date")
  rolling_stats = group[columns].rolling(3, closed='left').mean()
  group[new_columns] = rolling_stats
  group = group.dropna(subset=new_columns)
  return group 
```

```python
columns = ["GF", "GA", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]
new_columns = [f"{c}_rolling" for c in columns]
new_columns
```

```python
rolling_averages(group, columns, new_columns)
```

```python
matches_rolling = matches.groupby("Team").apply(lambda x:rolling_averages(x, columns, new_columns) )
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling 
```

```python
def make_predicitons(data, predictors):
  train = data[data["Date"] < '2022-01-01']
  test = data[data["Date"] > '2022-01-01']
  rf.fit(train[predictors], train["target"])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
  precision = precision_score(test["target"], preds)
  return combined,precision
  
```

```python
combined, precision = make_predicitons (matches_rolling, predictors + new_columns)
combined
```
